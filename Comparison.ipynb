```python
# Import libraries
import pandas as pd
```

```python
file_orig = 'school_summary_orig.csv'
file_ths = 'school_summary_ths.csv'

school_summary_orig = pd.read_csv(file_orig)
school_summary_ths = pd.read_csv(file_ths)
```

```python
school_summary_ths.index.name = None
school_summary_orig.index.name = None
school_summary_ths = school_summary_ths.rename(columns = {'Unnamed: 0': 'School Name'})
school_summary_orig = school_summary_orig.rename(columns = {'Unnamed: 0': 'School Name'})
```

```python
school_summary_vs = school_summary_orig.merge(school_summary_ths, on = ['School Name', 'School Type', 'Total Students','Total School Budget', 'Per Student Budget'], how = 'left', suffixes = ('_orig', '_ths' ))
school_summary_vs = school_summary_vs.set_index('School Name')
school_summary_vs_clean = school_summary_vs.loc[:,'Average Math Score_orig': '% Overall Passing_ths']
```

```python
school_summary_vs_clean['Avg Math vs Orig'] = school_summary_vs_clean['Average Math Score_ths']-school_summary_vs_clean['Average Math Score_orig']
school_summary_vs_clean['Avg Reading vs Orig'] = school_summary_vs_clean['Average Reading Score_ths']-school_summary_vs_clean['Average Reading Score_orig']
school_summary_vs_clean['% Passing Math vs Orig'] = school_summary_vs_clean['% Passing Math_ths']-school_summary_vs_clean['% Passing Math_orig']
school_summary_vs_clean['% Passing Reading vs Orig'] = school_summary_vs_clean['% Passing Reading_ths']-school_summary_vs_clean['% Passing Reading_orig']
school_summary_vs_clean['% Overall Passing vs Orig'] = school_summary_vs_clean['% Overall Passing_ths']-school_summary_vs_clean['% Overall Passing_orig']
school_summary_vs_ready = school_summary_vs_clean.loc[:,'Avg Math vs Orig':'% Overall Passing vs Orig']
```

```python
# Formatting
school_summary_vs_ready['Avg Math vs Orig'] = school_summary_vs_ready['Avg Math vs Orig'].map('{:,.2f}'.format)
school_summary_vs_ready['Avg Reading vs Orig'] = school_summary_vs_ready['Avg Reading vs Orig'].map('{:,.2f}'.format)
school_summary_vs_ready['% Passing Math vs Orig'] = school_summary_vs_ready['% Passing Math vs Orig'].map('{:,.2f}'.format)
school_summary_vs_ready['% Passing Reading vs Orig'] = school_summary_vs_ready['% Passing Reading vs Orig'].map('{:,.2f}'.format)
school_summary_vs_ready['% Overall Passing vs Orig'] = school_summary_vs_ready['% Overall Passing vs Orig'].map('{:,.2f}'.format)
school_summary_vs_ready
```
